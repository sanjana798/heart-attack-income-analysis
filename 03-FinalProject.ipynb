## Abstract
We set out to examine whether economic conditions help explain why some California counties experience higher heart disease mortality than others. Specifically, we focused on two county-level socioeconomic indicators—median household income and unemployment rate—and asked how they relate to age-adjusted heart disease mortality among adults aged 35+ in 2019–2021. This question is motivated by prior work on social determinants of health, which suggests that economic disadvantage (through limited access to care, chronic stress, and fewer health-promoting resources) should be associated with worse cardiovascular outcomes.

To answer this, we combined two public datasets. From the CDC/NCHS Heart Disease Mortality dataset, we filtered to California counties, kept only county-level observations for heart disease mortality, and averaged across demographic strata to obtain a single age-adjusted mortality rate per county (3-year smoothed around 2020). From the USDA ERS County-Level Data, we extracted unemployment rates for 2019–2021 and the most recent available median household income (2022), then reshaped the data so each county had one row with unemployment in 2019, 2020, 2021, and income. We merged the two data sources on county FIPS codes to create our analysis dataset.

Our exploratory data analysis included scatterplots with regression lines and a correlation matrix. Median household income showed a clear negative relationship with heart disease mortality—richer counties tended to have lower age-adjusted mortality rates. Unemployment was positively associated with mortality in all three years, with the strongest relationship in 2021. Simple linear regressions confirmed that both lower income and higher unemployment, considered separately, were associated with higher mortality, with income explaining more variance than any individual unemployment year.

Finally, we fit a multiple linear regression model with mortality as the outcome and both median household income and 2021 unemployment as predictors. The model captured a meaningful share of the cross-county variation in mortality, with income having a negative coefficient and unemployment a positive one. Residual plots suggested no major violations of linearity or homoscedasticity, and multicollinearity between predictors appeared limited. Overall, our findings support the conclusion that, within California, economic disadvantage at the county level is associated with higher heart disease mortality—but we emphasize that these are ecological associations, not causal estimates, and that unmeasured confounders and within-county inequalities remain important limitations.

## Research Question

How do county-level socioeconomic factors—specifically median household income and unemployment rate—relate to age-adjusted heart disease mortality among adults aged 35 and older across California counties between 2019 and 2021? In particular, we examine whether counties with lower income and higher unemployment experience significantly higher cardiovascular mortality using correlation and regression analysis.

## Background and Prior Work


Cardiovascular disease (CVD) is the leading cause of death in the United States and remains a major public health burden despite advances in prevention and treatment. Substantial geographic variation in heart disease mortality across U.S. counties suggests that factors beyond individual clinical risk—particularly social and economic conditions—play a major role in shaping cardiovascular outcomes. Prior research has established that social determinants of health (SDoH), including income, employment, education, neighborhood conditions, and access to healthcare, significantly influence both the development and progression of cardiovascular disease. The American Heart Association emphasizes that economic stability is a key upstream driver of cardiovascular risk, affecting diet, healthcare access, stress exposure, and long-term disease prognosis.

Recent peer-reviewed studies further quantify these relationships. Xia et al. (2024) show that area-level socioeconomic disadvantage is independently associated with increased cardiovascular events and mortality even after adjusting for traditional medical risk factors. Similarly, Powell-Wiley et al. (2022) highlight how persistent economic insecurity contributes to chronic stress and reduced preventive care utilization, both of which elevate long-term cardiovascular risk. Economic consequences also move in the opposite direction: Luo et al. (2023) estimate that heart disease and stroke are associated with billions of dollars in lost labor income nationally, reinforcing the bidirectional relationship between economic hardship and cardiovascular morbidity.

California provides a particularly important setting for examining these patterns because of its wide economic heterogeneity across counties. Statewide public health surveillance from the California Department of Public Health shows consistently higher cardiovascular disease prevalence


## Hypothesis


We hypothesize that California counties with lower median household income and higher unemployment rates will exhibit higher age-adjusted heart disease mortality among adults aged 35 and older. We specifically predict:

1. A **negative association** between median household income and heart disease mortality (higher income → lower mortality).
2. A **positive association** between unemployment rate and heart disease mortality (higher unemployment → higher mortality).
3. That these associations will remain statistically meaningful when income and unemployment are included together in a multiple linear regression model.

### Multiple Linear Regression: Socioeconomic Predictors of Heart Disease Mortality

Using the cleaned and merged dataset created during the Data and EDA checkpoints, we quantify how **median household income** and **county-level unemployment in 2021** relate to **age-adjusted heart disease mortality** across California counties. The EDA showed a roughly linear negative association between income and mortality and a positive association between unemployment and mortality, motivating a multiple linear regression framework.

We fit a multiple linear regression model with **heart disease mortality as the outcome variable** and **median household income and 2021 unemployment rate as predictors**. This model allows us to interpret the **independent contribution of each socioeconomic factor** while holding the other constant.


```python
import pandas as pd
import statsmodels
import statsmodels.formula.api as smf

from IPython.display import Image, display
```

```python
df_merged = pd.read_csv("/Users/uddhavs/Group127_FA25/data/df_merged.csv")
df_merged.head()
```

```python
summary_vars = [
    "mortality_rate",
    "median_household_income",
    "unemp_2021"
]

df_merged[summary_vars].describe()
```


The summary statistics show substantial variation across California counties in both **economic conditions** and **heart disease mortality**. Mortality rates span a wide range, as do median household incomes and unemployment rates. This heterogeneity is necessary for identifying meaningful statistical associations in the regression model.


```python
# %%
# Fit the multiple linear regression model

final_model = smf.ols(
    "mortality_rate ~ median_household_income + unemp_2021",
    data=df_merged
).fit()

final_model.summary()
```

The regression results indicate that **median household income is negatively associated with heart disease mortality**, while **unemployment is positively associated with mortality**, holding the other predictor constant. This aligns with our hypothesis and prior public-health literature on social determinants of cardiovascular disease.

The model’s R-squared value indicates that a meaningful portion of the between-county variation in heart disease mortality is explained by these two socioeconomic predictors alone. While the model is intentionally simple and does not capture all clinical or demographic risk factors, it provides strong evidence that **economic conditions are significantly related to cardiovascular mortality at the county level**.


```python
# Display the saved residuals vs fitted values plot from EDA 

display(Image("results/residuals.png"))
```

**Figure 5.** Residuals vs. fitted values for the multiple linear regression model.

The residual–fitted plot shows that residuals are approximately centered around zero with no strong curvature or funneling pattern, supporting the assumptions of **linearity** and **roughly constant variance**. While some dispersion is expected due to omitted variables (such as healthcare access, comorbidities, and behavioral risk factors), there is no strong visual evidence of severe model misspecification. This supports the validity of the regression-based conclusions.


### Robustness Check: Using 2020 Unemployment Instead of 2021

To assess whether our main findings are sensitive to the specific choice of unemployment year, we perform a robustness check by re-estimating the regression model using **2020 unemployment** instead of 2021. This is especially important because 2020 reflects peak labor-market disruption during the COVID-19 pandemic, which may have a different relationship with mortality than 2021.

We fit a second multiple linear regression model with:
- **Outcome:** Age-adjusted heart disease mortality  
- **Predictors:** Median household income and 2020 unemployment rate  

We then compare the **signs, magnitudes, and statistical significance** of the coefficients to those from the primary model.


```python
robust_model_2020 = smf.ols(
    "mortality_rate ~ median_household_income + unemp_2020",
    data=df_merged
).fit()

robust_model_2020.summary()
```

The robustness regression using **2020 unemployment** yields coefficient signs that are consistent with the primary 2021-based model. Median household income remains **negatively associated** with heart disease mortality, while unemployment remains **positively associated** with mortality across counties.

The similarity of coefficient directions and overall model fit indicates that our core conclusion—that **economic disadvantage is associated with higher heart disease mortality at the county level**—is **not driven by a single year of unemployment measurement**. This strengthens the credibility and generalizability of our findings.


```python
comparison = pd.DataFrame({"2021 Model": final_model.params,"2020 Model": robust_model_2020.params})
comparison
```

The table above compares coefficient estimates from the primary model (using **2021 unemployment**) and the robustness model (using **2020 unemployment**). In both specifications, **median household income retains a negative coefficient of similar magnitude** (−0.001002 vs. −0.001065), indicating a stable inverse relationship between income and heart disease mortality across model choices. 

Unemployment coefficients are **positive in both models** (4.19 for 2021 unemployment and 2.89 for 2020 unemployment), showing that higher unemployment is consistently associated with higher mortality. Although the magnitudes differ—reflecting distinct labor market conditions during the pandemic—the direction and substantive interpretation remain the same. 

Overall, this robustness check confirms that the observed association between **county-level economic conditions and heart disease mortality** is **not driven by the choice of a single unemployment year**, strengthening confidence in the reliability of our primary findings.


### Model Diagnostics and Assumption Checks

In addition to estimating regression coefficients, we directly evaluated whether the linear model assumptions were reasonably satisfied. We examined the **residuals vs. fitted values plot** (Figure 5) to check for nonlinearity and heteroscedasticity. The residuals were approximately centered around zero with no strong curvature or funneling pattern, supporting the assumptions of **linearity and constant variance**.

We also assessed **multicollinearity** between predictors using the Variance Inflation Factor (VIF). Median household income and unemployment exhibited low VIF values, indicating that multicollinearity was not a concern and that coefficient estimates are statistically stable. These diagnostic checks strengthen confidence that the regression results reflect genuine county-level associations rather than violations of modeling assumptions.

Together with the robustness check using 2020 unemployment, these diagnostics confirm that our findings are not artifacts of poor model specification or numerical instability.

## Discussion and Conclusion

In this study, we examined how county-level socioeconomic conditions relate to age-adjusted heart disease mortality among adults aged 35 and older across California counties. Using a multiple linear regression framework, we found that **median household income is negatively associated with heart disease mortality**, while **unemployment is positively associated with mortality**. These relationships were first suggested by our exploratory visualizations and then confirmed through regression modeling. A robustness check using 2020 instead of 2021 unemployment produced substantively consistent results, indicating that the observed associations are not driven by a single year of labor-market conditions.

These findings align closely with the prior literature reviewed in our Background section. Previous studies and public health reports emphasized the importance of social determinants of health—particularly economic stability—in shaping cardiovascular outcomes. Our results extend this work by providing a **California-specific, county-level validation** of these relationships using recent post-pandemic-era data. The consistency between our empirical findings and the theoretical framework of social determinants of health strengthens confidence that economic disadvantage operates as a meaningful upstream factor in cardiovascular risk.

Taken together, the results paint a clear picture: **counties with lower income and higher unemployment tend to experience higher heart disease mortality**. This suggests that economic conditions may influence cardiovascular health not only through individual behaviors, but also through broader structural mechanisms such as access to preventive care, healthcare infrastructure, chronic stress, food environments, and environmental exposures. While our model does not establish causality, the consistent direction and magnitude of these associations across analyses strongly support the public health relevance of economic context.

We did not observe any direct contradictions between our results and the existing literature; however, the differing magnitudes of the unemployment coefficients between the 2020 and 2021 models highlight how **macroeconomic shocks**, such as the COVID-19 pandemic, may alter the strength of economic–health relationships over time. This variability reflects the complex and dynamic nature of economic stressors and suggests that the relationship between unemployment and mortality is not fixed, but may depend on broader historical and policy contexts.

Several limitations are important to acknowledge. First, our data are **aggregate at the county level**, which introduces the risk of **ecological fallacy**—we cannot infer that individuals with lower income or who are unemployed necessarily face higher personal risk based solely on these results. Second, the analysis is fundamentally **cross-sectional**, limiting our ability to assess causal pathways or temporal dynamics. Third, our model includes only two socioeconomic predictors, omitting potentially important confounders such as race and ethnicity composition, healthcare access, insurance coverage, physical activity, diet, and environmental exposures. Additionally, the mortality outcomes are based on **age-adjusted, spatially smoothed estimates**, which improve stability for small counties but may obscure localized heterogeneity.

Future work could substantially extend this analysis. A logical next step would be to incorporate **additional social determinants** such as education, poverty rate, rural–urban classification, healthcare provider density, and insurance coverage to better isolate independent effects. A **longitudinal panel analysis** using multiple years of mortality and economic data would allow for stronger causal inference and the study of lagged economic effects. Finally, integrating **individual-level datasets**—where ethically and legally permissible—could help bridge the gap between population-level associations and individual risk.

In conclusion, our study provides clear evidence that **economic conditions are strongly associated with heart disease mortality at the county level in California**. These findings reinforce the importance of addressing socioeconomic inequality as part of comprehensive cardiovascular disease prevention strategies. Public health interventions that target not only clinical risk factors but also upstream economic determinants may be essential for reducing persistent geographic disparities in heart disease outcomes.
